# **TOP 15 STATES WITH MOST PROFITS**

```python
dataset["State"].unique()
```

```python
state_profit=pd.DataFrame(dataset.groupby("State")["Profit"].sum())
state_profit.reset_index(inplace=True)
```

**Here a new dataframe(table) is created named as state_profit , which are grouped by state and the profit is calculated by each state  then in the second cell it is sorted in ascending order**

```python
state_profit=state_profit.sort_values(by="Profit",ascending=False)
state_profit.head(15)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 6))
plt.bar(state_profit["State"], state_profit["Profit"],color='blue')
plt.xlabel("State")
plt.ylabel("Total Profit ")
plt.title("Total profit by state")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# MOST PROFITABLE CATEGORY IN THE DATASET

```python
dataset["Category"].unique()
```

```python
dataset["Sub-Category"].unique()
```

```python
categories_info=pd.DataFrame(dataset.groupby("Category")["Profit"].sum())
categories_info.reset_index(inplace=True)
categories_profit=categories_info.sort_values(by="Profit",ascending=False)
categories_profit.head(3)
```

```python
import matplotlib.pyplot as plt
plt.bar(categories_profit["Category"],categories_profit["Profit"])
plt.xticks(categories_profit['Category'] , rotation = "vertical")
plt.xlabel("Categories")
plt.ylabel("Profit")
plt.title('Category-Wise Profit')
plt.show()
```

# CATEGORIES SOLD THE MOST

```python
dataset['Category'].value_counts()
```

```python
cat_sales=dataset.groupby("Category").sum()
cat_sales.reset_index(inplace=True)
cat_sales=cat_sales.sort_values(by="Sales",ascending=False)
cat_sales.head(5)
```

```python
plt.bar(cat_sales["Category"], cat_sales['Sales'],color='red')
plt.title('Category-most-popular')
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()
```

# CITY HAVING HIGHEST SALES

```python
city_highest_sales = pd.DataFrame(dataset.groupby('City')['Sales'].sum())
city_highest_sales =city_highest_sales.sort_values(by="Sales",ascending=False)
city_highest_sales.head()
```

```python
plt.bar(city_highest_sales.index[0:20], city_highest_sales['Sales'][0:20],color="purple")
plt.xticks(rotation = 90)
plt.xlabel("City" )
plt.ylabel("Sales")
plt.title('CITY-MOST-SALES')
plt.show()
```
